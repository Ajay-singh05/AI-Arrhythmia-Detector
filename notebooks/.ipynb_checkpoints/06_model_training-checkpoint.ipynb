```python
 # ============================================
# Notebook 06: Model Training (All Records)
# ============================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import joblib  # For saving the model
import os      # For folder management

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression

# -------------------------------
# 1. Load the Big Dataset
# -------------------------------
try:
    df = pd.read_csv("combined_ecg_features.csv")
    print("\nüìä Dataset Loaded Successfully")
    print(f"   Total Samples: {len(df)}")
    print(f"   Columns: {df.columns.tolist()}")
except FileNotFoundError:
    print("‚ùå Error: 'combined_ecg_features.csv' not found. Please run Notebook 04 first.")
    df = pd.DataFrame()

if not df.empty:
    # -------------------------------
    # 2. Data Cleaning
    # -------------------------------
    # Remove 'record_id' as it's not a feature for prediction
    if "record_id" in df.columns:
        df = df.drop(columns=["record_id"])

    # Remove rows with empty values
    df = df.dropna().reset_index(drop=True)

    # -------------------------------
    # 3. Define Target (The Goal)
    # -------------------------------
    # Logic: If Heart Rate > 100, we label it as "1" (Abnormal/Tachycardia)
    # You can adjust this threshold if needed
    df["target"] = np.where(df["Heart_Rate(bpm)"] > 100, 1, 0)

    # Check how many Normal vs Abnormal we have
    counts = df["target"].value_counts()
    print("\nTarget Distribution:")
    print(f"   Normal (0):   {counts.get(0, 0)}")
    print(f"   Abnormal (1): {counts.get(1, 0)}")

    # -------------------------------
    # 4. Prepare Features (X) and Target (y)
    # -------------------------------
    # Select only the numerical features used for training
    feature_cols = [
        "RR_prev(s)",
        "RR_next(s)",
        "QRS_duration(s)",
        "R_amp",
        "QRS_max",
        "QRS_min"
    ]
    
    # Check if all columns exist
    existing_cols = [c for c in feature_cols if c in df.columns]
    X = df[existing_cols]
    y = df["target"]

    # -------------------------------
    # 5. Split Data (Train vs Test)
    # -------------------------------
    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=0.25,      # 25% data for testing
        random_state=42,
        stratify=y           # Keeps the % of abnormal beats same in train and test
    )

    # -------------------------------
    # 6. Scaling (Standardization)
    # -------------------------------
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # -------------------------------
    # 7. Train the Model (THE UPGRADE)
    # -------------------------------
    print("\n‚öôÔ∏è Training Model on All Records...")
    
    # We use class_weight='balanced' to handle the imbalance 
    # (since we have way more normal beats than abnormal ones)
    model = LogisticRegression(max_iter=2000, class_weight='balanced')
    model.fit(X_train_scaled, y_train)
    
    print("‚úÖ Training Completed!")

    # -------------------------------
    # 8. Evaluation
    # -------------------------------
    y_pred = model.predict(X_test_scaled)
    accuracy = accuracy_score(y_test, y_pred)

    print(f"\nüéØ Test Accuracy: {accuracy:.4f}")
    
    print("\nüìã Classification Report:")
    try:
        from IPython.display import display
        report = classification_report(y_test, y_pred, output_dict=True)
        display(pd.DataFrame(report).T.round(2))
    except:
        print(classification_report(y_test, y_pred))

    # -------------------------------
    # 9. Visualize Confusion Matrix
    # -------------------------------
    cm = confusion_matrix(y_test, y_pred)
    
    plt.figure(figsize=(5, 4))
    plt.imshow(cm, cmap="Blues", alpha=0.8)
    plt.colorbar()
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title("Confusion Matrix")
    plt.xticks([0, 1], ["Normal", "Abnormal"])
    plt.yticks([0, 1], ["Normal", "Abnormal"])
    
    # Add numbers inside the boxes
    for i in range(2):
        for j in range(2):
            plt.text(j, i, cm[i, j], ha="center", va="center", color="red")
            
    plt.show()

    # -------------------------------
    # 10. SAVE EVERYTHING
    # -------------------------------
    if not os.path.exists('models'):
        os.makedirs('models')
        
    joblib.dump(model, 'models/rf_ecg_model.pkl')
    joblib.dump(scaler, 'models/scaler.pkl')

    print("\nüíæ SUCCESS: 'rf_ecg_model.pkl' and 'scaler.pkl' saved in 'models/' folder.")
```

```python
import pickle
import sklearn

artifact = {
    "model": model,
    "scaler": scaler,
    "sklearn_version": sklearn.__version__
}

with open("ecg_logreg_model.pkl", "wb") as f:
    pickle.dump(artifact, f)

print("Model saved with sklearn version:", sklearn.__version__)
```
