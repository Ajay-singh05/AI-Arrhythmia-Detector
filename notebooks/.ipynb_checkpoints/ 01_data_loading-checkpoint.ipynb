```python
 # ===============================
# Notebook 01: Load & Inspect ECG
# ===============================

import wfdb
import matplotlib.pyplot as plt
import numpy as np

# Load ECG Record and Annotations
record = wfdb.rdrecord('100', pn_dir='mitdb')
annotation = wfdb.rdann('100', 'atr', pn_dir='mitdb')

# Extract single lead (MLII)
ecg_signal = record.p_signal[:, 0]
fs = record.fs
time = np.arange(len(ecg_signal)) / fs

# -------------------------------
# Plot 1: Raw ECG Signal
# -------------------------------
plt.figure(figsize=(14, 4))
plt.plot(time[:10*fs], ecg_signal[:10*fs], linewidth=1)
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude (mV)")
plt.title("Raw ECG Signal (Record 100, Lead MLII)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# -------------------------------
# Plot 2: ECG with R-Peak Annotations
# -------------------------------
r_peaks = annotation.sample
mask = r_peaks < 10 * fs

plt.figure(figsize=(14, 4))
plt.plot(time[:10*fs], ecg_signal[:10*fs], linewidth=1, label="ECG Signal")
plt.scatter(r_peaks[mask] / fs,
            ecg_signal[r_peaks[mask]],
            s=25,
            label="R-peaks")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude (mV)")
plt.title("ECG Signal with Annotated R-peaks (First 10 seconds)")
plt.legend(loc="upper right")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
