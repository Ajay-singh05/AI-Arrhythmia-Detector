```python
 # ===============================
# Notebook 02: ECG Signal Preprocessing
# ===============================

import wfdb
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt

# -------------------------------
# Load ECG Record & Annotations
# -------------------------------
record = wfdb.rdrecord('100', pn_dir='mitdb')
annotation = wfdb.rdann('100', 'atr', pn_dir='mitdb')

ecg_signal = record.p_signal[:, 0]
fs = record.fs
time = np.arange(len(ecg_signal)) / fs

# -------------------------------
# Bandpass Filter (0.540 Hz)
# -------------------------------
def bandpass_filter(signal, lowcut=0.5, highcut=40, fs=360, order=2):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, signal)

ecg_filtered = bandpass_filter(ecg_signal, fs=fs)

# -------------------------------
# Baseline Handling
# -------------------------------
# Baseline wander already removed by high-pass filtering
ecg_clean = ecg_filtered

# -------------------------------
# Z-score Normalization (SAFE)
# -------------------------------
ecg_normalized = (ecg_clean - np.mean(ecg_clean)) / np.std(ecg_clean)

# -------------------------------
# Plot 1: Preprocessed ECG
# -------------------------------
plt.figure(figsize=(14, 4))
plt.plot(time[:10*fs], ecg_normalized[:10*fs], linewidth=1.2)
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude (z-score)")
plt.title("Preprocessed ECG Signal (0.540 Hz Bandpass)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# -------------------------------
# Plot 2: ECG with R-peaks (CLEAR)
# -------------------------------
r_peaks = annotation.sample
mask = r_peaks < 10 * fs

plt.figure(figsize=(14, 4))
plt.plot(time[:10*fs],
         ecg_normalized[:10*fs],
         linewidth=1.2,
         label="ECG")

plt.scatter(r_peaks[mask] / fs,
            ecg_normalized[r_peaks[mask]],
            s=40,
            marker='o',
            label="R-peaks")

plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude (z-score)")
plt.title("Preprocessed ECG with Annotated R-peaks")
plt.legend(loc="upper right")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
